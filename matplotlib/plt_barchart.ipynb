```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

BAR PLOT ------------------------------------------

```python
x=["Python", "Java", "C++", "JavaScript", "Ruby"]
y=[50, 40, 30, 20, 10]

plt.bar(x,y,width=.4,color="black")

plt.xlabel("Programming Languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("My first bar chart") 


```

```python
x=["Python", "Java", "C++", "JavaScript", "Ruby"]
y=[50, 40, 30, 20, 10]

c=["g","r","b","r","r"]
plt.bar(x,y,width=.4,color=c)

plt.xlabel("Programming Languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("My first bar chart")




```

```python
x=["Python", "Java", "C++", "JavaScript", "Ruby"]
y=[50, 40, 30, 20, 10]

c=["g","r","b","r","r"]
plt.bar(x,y,width=.4,color=c,edgecolor="black", linewidth=2,alpha=.5)

plt.xlabel("Programming Languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("My first bar chart")
```

```python
x=["Python", "Java", "C++", "JavaScript", "Ruby"]
y=[50, 40, 30, 20, 10]

c=["g","r","b","r","r"]
plt.bar(x,y,width=.4,color=c,label="PLOT LABLE")

plt.xlabel("This is X Label",fontsize=15)
plt.ylabel("This is Y Label",fontsize=15)
plt.title("My first bar chart:Title")

plt.legend(loc="upper right", fontsize=15) #label=legend
```

Creating Multiple Bar Charts................................................

```python
#Composing 2 bar charts in single one.

x=["Python", "Java", "C++", "JavaScript", "Ruby"]
y=[85, 70, 60, 82, 10]

x1=["Python", "Java", "C++", "JavaScript", "Ruby"]
y1=[10, 20, 30, 40, 50]


plt.bar(x,y,width=.4,color="yellow",label="PLOT LABLE")
plt.bar(x1,y1,width=.4,color="red",label="PLOT LABLE 2")


plt.xlabel("This is X Label",fontsize=15)
plt.ylabel("This is Y Label",fontsize=15)
plt.title("My first bar chart:Title")

plt.legend(loc="upper center", fontsize=10) #label=legend


```

```python
#Plotting bar side by side


x=["Python", "Java", "C++", "JavaScript", "Ruby"]
y=[85, 70, 60, 82, 10]


y1=[10, 20, 30, 40, 50]

p=np.arange(len(x)) #position of bars 
p1=[j+0.4 for j in p] #position of bars


plt.bar(p,y,width=.4,color="yellow",label="PLOT LABLE")
plt.bar(p1,y1,width=.4,color="red",label="PLOT LABLE 2")


plt.xlabel("This is X Label",fontsize=15)
plt.ylabel("This is Y Label",fontsize=15)
plt.title("My first bar chart:Title")

plt.legend(loc="upper center", fontsize=10) #label=legend


```

```python
#More conventional way of plotting bar charts side by side, i added 0.4 to the position of the first bar chart to make it side by side
years =np.arange(2012,2015)

value1=[10,20,30]
value2=[20,30,40]

plt.bar(years,value1,width=.2,color="yellow",label="PLOT LABLE")
plt.bar(years+0.4,value2,width=.2,color="red",label="PLOT LABLE 2")
plt.xlabel("This is X Label",fontsize=15)
plt.ylabel("This is Y Label",fontsize=15)

```
